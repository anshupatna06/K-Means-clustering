```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generate Dataset:-
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)
```

```python
# K-Means Helper Functions
# ------------------------------
def initialize_centroids(X, k):
    np.random.seed(42)
    random_idx = np.random.permutation(X.shape[0])[:k]
    return X[random_idx]

def assign_clusters(X, centroids):
    distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)  
    return np.argmin(distances, axis=1)

def update_centroids(X, labels, k):
    new_centroids = []
    for j in range(k):
        cluster_points = X[labels == j]
        if len(cluster_points) > 0:
            new_centroids.append(cluster_points.mean(axis=0))
        else:
            # Reinitialize centroid randomly if cluster is empty
            new_centroids.append(X[np.random.randint(0, X.shape[0])])
    return np.array(new_centroids)
    

def compute_sse(X, labels, centroids):
    # Sum of squared distances of points to their assigned centroid
    sse = 0
    for i in range(len(centroids)):
        cluster_points = X[labels == i]
        sse += np.sum((cluster_points - centroids[i]) ** 2)
    return sse
```

```python
def compute_sse(X, labels, centroids):
    sse = 0
    for i in range(len(centroids)):
        cluster_points = X[labels == i]
        sse += np.sum((cluster_points - centroids[i]) ** 2)
    return sse
```

```python
# K-Means Algorithm
# ------------------------------
def kmeans(X, k, max_iters=100, tol=1e-4):
    centroids = initialize_centroids(X, k)

    for i in range(max_iters):
        labels = assign_clusters(X, centroids)
        new_centroids = update_centroids(X, labels, k)

        # Convergence check
        if np.all(np.abs(new_centroids - centroids) < tol):
            break
        centroids = new_centroids
    
    return centroids, labels
```

```python
# Run Scratch KMeans:-
k = 4
scratch_centroids, scratch_labels = kmeans(X, k)
scratch_sse = compute_sse(X, scratch_labels, scratch_centroids)

plt.scatter(X[:, 0], X[:, 1], c=scratch_labels, cmap='viridis', s=30)
plt.scatter(scratch_centroids[:, 0], scratch_centroids[:, 1], c='red', marker='X', s=200, label="Scratch Centroids")
plt.title("K-Means (Scratch Implementation)")
plt.legend()
plt.show()
```

```python
# Elbow Method:-
sse_values = []
K = range(1, 11)  # test k = 1 to 10

for k in K:
    centroids, labels = kmeans(X, k)
    sse = compute_sse(X, labels, centroids)
    sse_values.append(sse)

# Plot Elbow Curve
plt.plot(K, sse_values, 'bx-')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("SSE (Sum of Squared Errors)")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
# Sklearn KMeans:-
# ------------------------------
from sklearn.cluster import KMeans
sk_model = KMeans(n_clusters=k, random_state=42, n_init=10)
sk_labels = sk_model.fit_predict(X)
sk_centroids = sk_model.cluster_centers_
sk_sse = sk_model.inertia_

plt.scatter(X[:, 0], X[:, 1], c=sk_labels, cmap='viridis', s=30)
plt.scatter(sk_centroids[:, 0], sk_centroids[:, 1], c='red', marker='X', s=200, label="Sklearn Centroids")
plt.title("K-Means (Sklearn)")
plt.legend()
plt.show()
```

```python
# Print Comparison:-
print("Scratch KMeans SSE:", scratch_sse)
print("Sklearn KMeans SSE:", sk_sse)

print("\nScratch Centroids:\n", scratch_centroids)
print("\nSklearn Centroids:\n", sk_centroids)
```
